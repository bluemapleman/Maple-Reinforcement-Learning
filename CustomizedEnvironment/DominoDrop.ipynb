# DominoDrop

- observation_space：

12*4网格，每个方格对应一个颜色方格或空

方块颜色编码：{黑色：1，红色：2，绿色：3，橘色：4，蓝色：5，黄色：6，白色：7（特殊处理）}

- action_space

竖直方块：(0,1,2,3)四个列，选择一列下落

水平方块：（0，1，2）三个列，选择一列下落

- done

无法再往网格里下落给定的矩形

- reward

同色方块碰撞时消除，r=n*100

若产生第二次连锁同色碰撞消除，r=n*500

第三次连锁, r=n*1000

...

第i次连锁(i>1)，r=n*(500*(i-1))
